```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
errors = []
for i in range(0,30,2):
    print(i)
    errors.append(np.linspace(i,i+2,50))
```

```python
errors
```

```python
se = np.square(errors)
```

```python
mse = []
for a in se:
    mse.append(np.mean(a))
```

```python
mse
```

```python
plt.plot(mse)
```

```python
rmse = []
for a in mse:
    rmse.append(np.sqrt(a))
rmse
```

```python
plt.plot(rmse)
```

```python
lse = []
for a in se:
    lse.append(np.log(a))
lse
```

```python
mlse = []
for a in lse:
    mlse.append(np.mean(a))
mlse
```

```python
plt.plot(mlse)
```
